```python
from sympy import *
from scipy.misc import derivative

x = symbols('x')
```

### Уровень 0

```python
# Найти производную и вторую производную f(x) = x**5 + 4*sin(2*x) + cos(3*x+3) в точке х=1
# f'(x)=5x**4+4*2*cos(2x)-3sin(3x+3)=5+8cos(2)-3sin(6) = 2.509

def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x+3)

f_dif1 = derivative(f, x0=1.0, n=1, dx=1e-6)
f_dif2 = derivative(f, x0=1.0, n=2, dx=1e-6)

print('Первая производная функции в точке х=1 равна ', f_dif1)
print('Вторая производная функции в точке х=1 равна ', f_dif2)
```

### Уровень 1

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Построить график первой и второй производной функции f(x)=sin**5(2x+1) в интервале [-5;5]

def F(x):
    return (sin(2*x+1))**5

F_diff1 = []
F_diff2 = []
N = []

for n in np.linspace(-5, 5, 100, endpoint = True):
    fd = derivative(F, x0=n, n=1, dx=1e-6)
    F_diff1.append(fd)
    fd = derivative(F, x0=n, n=2, dx=1e-6)
    F_diff2.append(fd)
    N.append(n)
```

```python
plt.plot(N, F_diff1, label="f'(x)")
plt.plot(N, F_diff2, label="f''(x)")
plt.xlabel("X")
plt.ylabel("f'(x), f''(x)")
plt.legend()
plt.grid()
plt.xlim([-5,5])
plt.show()
```
